```python
import numpy as np
import scipy.stats as stats #new library
import matplotlib.pyplot as plt
```

```python
# Step 1: Generate Dataset
k = 51
np.random.seed(0)
data = 60 + np.random.randint(0, 15, size=12) + (k % 5)
print("Dataset (exam scores):", data)

# Step 2: State the Hypothesis
population_mean = 65

# Step 3: Perform One-Sample t-test
t_statistic, p_value = stats.ttest_1samp(data, population_mean)

# Step 4: Compute 95% Confidence Interval
sample_mean = np.mean(data)
sample_sem = stats.sem(data)  # standard error
df = len(data) - 1
ci = stats.t.interval(0.95, df=df, loc=sample_mean, scale=sample_sem)

# Step 5: Print Results
print(f"\nSample Mean: {sample_mean:.2f}")
print(f"Hypothesized Population Mean: {population_mean}")
print(f"t-Statistic: {t_statistic:.2f}")
print(f"Degrees of Freedom: {df}")
print(f"p-Value: {p_value:.4f}")
print(f"95% Confidence Interval: [{ci[0]:.2f}, {ci[1]:.2f}]")
```

```python
# Step 6: Decision Rule
alpha = 0.05
if p_value < alpha:
    print(f"Decision: Reject H0. The sample mean is significantly different from {population_mean}.")
else:
    print(f"Decision: Fail to reject H0. The sample mean is not significantly different from {population_mean}.")

```

```python
# Step 7: Visualization
plt.figure(figsize=(6,4))
plt.bar(1, sample_mean, color="white", label='Sample Mean')
plt.errorbar(1, sample_mean, yerr=sample_sem, fmt='k', capsize=8, label="Â±1 SE")
plt.axhline(y=population_mean, color='r', linestyle='--', linewidth=1.5, label='Hypothesized Mean (65)')

plt.title("Sample Mean vs Hypothesized Mean")
plt.xlim(0.5,1.5)
plt.ylabel("Exam Score")
plt.xticks([])
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
